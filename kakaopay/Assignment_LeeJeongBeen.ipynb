### Setting & Data Loading & 변수 생성

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 100

pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
# 데이터 로딩 1 (user_usage)
raw_uu = pd.read_csv("data/user_usage.csv")

print(raw_uu.shape)
raw_uu.head()
```

```python
raw_uu['user_type'].value_counts()
```

```python
# 데이터 로딩 2 (transaction)
raw_tx = pd.read_csv("data/transaction.csv")

print(raw_tx.shape)
raw_tx.head()
```

```python
# date변수 일자로 변환, 요일 변수 생성
raw_tx['date'] = pd.to_datetime(raw_tx['date'])
raw_tx['date(clean)'] = raw_tx['date'].dt.date
raw_tx['weekday'] = raw_tx['date'].dt.weekday_name
raw_tx.head()
```

```python
# 연령 grouping (10세 단위)
raw_tx.loc[raw_tx['age'] <= 19, "age(group)"] = "00 ~ 19"
raw_tx.loc[(raw_tx['age'] >= 20) & (raw_tx['age'] <= 29), "age(group)"] = "20 ~ 29"
raw_tx.loc[(raw_tx['age'] >= 30) & (raw_tx['age'] <= 39), "age(group)"] = "30 ~ 39"
raw_tx.loc[(raw_tx['age'] >= 40) & (raw_tx['age'] <= 49), "age(group)"] = "40 ~ 49"
raw_tx.loc[(raw_tx['age'] >= 50) & (raw_tx['age'] <= 59), "age(group)"] = "50 ~ 59"
raw_tx.loc[(raw_tx['age'] >= 60) & (raw_tx['age'] <= 69), "age(group)"] = "60 ~ 69"
raw_tx.loc[raw_tx['age'] >= 70, "age(group)"] = "70 ~ 99"

raw_tx[["age", "age(group)"]].head()
```

```python
# raw_tx의 null값 확인
raw_tx.loc[raw_tx['age'].isnull() == True]
```

```python
# raw_tx - raw_uu left join

merge_ing = pd.merge(raw_tx, raw_uu, on="user_id", how="left")

print(merge_ing.shape)
merge_ing.head()
```

### Q1) 주어진 두 데이터를 활용하여 탐색 후 1개 혹은 2개 가설을 수립하고 검정해주세요. <br>또한, 해당 가설을 수립한 이유와 탐색 과정을 기술해주세요.

### 가설 : 요일별로 거래추이에 다른 부분이 있을 것이다.
#### 가설 수립 이유 : 대부분의 고객들은 요일별 행동패턴이 다르기 때문에 데이터도 요일별로 다른 부분이 있을 것이라 생각하였다.<br>어느 요일에 많이 거래되고, 거래되지 않는지 확인하고, 그렇다면 그에 영향을 끼칠만한 factor은 어떤 것이 있을지 주어진 데이터를 베이스로 검증하고자 한다. 

```python
# 탐색 - 요일별 거래건수
table = pd.pivot_table(raw_tx,
                       index = 'weekday',
                       values = 'tid',
                       aggfunc = len,
                       )
weekday_columns = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
table = table.loc[weekday_columns]
table
```

```python
sns.barplot(data=raw_tx, x='weekday', y='tid',estimator=len, order=table.index, errwidth=0)
```

```python
# 탐색 - 요일별 거래액
table = pd.pivot_table(raw_tx,
                       index = 'weekday',
                       values = 'amount',
                       aggfunc = "sum"
                       )
weekday_columns = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
table = table.loc[weekday_columns]
table
```

```python
sns.barplot(data=raw_tx, x='weekday', y='amount',estimator=sum, order=table.index, errwidth=0)
```

##### → 거래량 및 거래액 모두 주말에 많고 화, 수요일에는 감소하는 추이를 보임

```python
# 그럼 요일 x 타 요인 별 거래에 차이가 있을까?
# 1. 성별
table = pd.pivot_table(raw_tx,
                       index = 'weekday',
                       columns = 'gender',
                       values = 'tid',
                       aggfunc = len
                       )
table
weekday_columns = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
table = table.loc[weekday_columns]
table
```

```python
sns.barplot(data=raw_tx, x='weekday', y='tid', hue='gender', estimator=len, order=table.index, errwidth=0)
```

##### → 1번 성별보다 2번 성별의 전체적인 거래비중이 약 50%가량 적을 뿐, 요일별 차이는 없는것으로 보임

```python
# 2. 연령 (10세 단위)
table = pd.pivot_table(raw_tx,
                       index = 'weekday',
                       columns = 'age(group)',
                       values = 'tid',
                       aggfunc = len
                       )
table
weekday_columns = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
table = table.loc[weekday_columns]
table
```

```python
plt.figure(figsize=[15,6])
sns.barplot(data=raw_tx, x='weekday', y='tid', hue='age(group)', estimator=len, order=table.index, errwidth=0)
```

##### → 30대 이상 그룹은 요일별 거래량에 차이가 있으나, 20대의 경우 차이가 크지 않음

```python
# 3. 성 x 연령 (10세 단위)
table = pd.pivot_table(raw_tx,
                       index = 'weekday',
                       columns = ['gender', 'age(group)'],
                       values = 'tid',
                       aggfunc = len
                       )
table
weekday_columns = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
table = table.loc[weekday_columns]
table
```

```python
table.style.background_gradient(cmap='summer', axis=0)
```

```python
table.style.background_gradient(cmap='summer', axis=1)
```

##### → 대부분의 거래는 1번 성별의 3,40대 고객에서 많이 일어남
##### → 보통 주말에 많이 거래가 일어나나, 1번 성별의 10대, 20대는 월요일에 많이 거래를 함
